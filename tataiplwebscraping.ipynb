```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
web = requests.get("https://www.iplt20.com/auction/2022")
web
```

```python
soup = BeautifulSoup(web.content, 'lxml')
soup
```

```python
soup.prettify()
```

```python
table1 = soup.find("table", class_="ih-td-tab w-100 auction-tbl")
table1
```

```python
columns = table1.find_all("th")
for i in columns:
  print(i.text)
```

```python
table_body = table1.find("tbody", id="pointsdata")
print(table_body)
```

```python
rows = table_body.find_all("tr")
sns = []
clubslist = []
price = []
overseas_player = []
total_players = []
for i in rows:
  # print(i.text)
  sn = i.find_all("td")
  clubs = i.find_all("h2")
  print(sn[0].text)
  sns.append(sn[0].text)
  # print(club.text)
  # clubs.append(club)
  for club in clubs:
    print(club.text)
    clubslist.append(club.text)
  print(sn[-3].text)
  price.append(sn[-3].text)
  print(sn[-2].text)
  overseas_player.append(sn[-2].text)
  print(sn[-1].text)
  total_players.append(sn[-1].text)
```

```python
data = {
    "Serial No." : sns,
    "Teams": clubslist,
    "Capital" : price,
    "Overseas Players":overseas_player,
    "Total Players" : total_players
}
df = pd.DataFrame(data)
df.set_index("Serial No.", inplace=True)
df
```

```python
df.to_csv("Ipldata.csv")
```
